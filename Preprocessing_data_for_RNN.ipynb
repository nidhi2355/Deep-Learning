```python
import numpy as np
docs=['go india', 'india india', 'hip hip hurray', 'jeetega bhyi jeetega india jeetega','bharat mata ki jai','kohli kohli', 'sachin sachin', 'dhoni dhoni', 'modi ji ki jai', 'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer= Tokenizer(oov_token='<nothing>') #at time of prediction if we get any word out of vocabulary, it will be replaced with this word
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
# transforming our input data into sequence of numbers
#word->number that denotes that word
sequences= tokenizer.texts_to_sequences(docs)
sequences
```

```python
#padding is making the size of these sequnces same by adding zeores either at start or end of sequences with smaller size
from tensorflow.keras.utils import pad_sequences
sequences= pad_sequences(sequences, padding='post') #end of sequence
```
